# Creating four different charts using python Programming language.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing first dataset
df_hotdog_winner=pd.read_excel("hotdog-contest-winners.xlsm")
df_hotdog_winner.head()
```

```python
# importing second datasets
df_hotdog_places=pd.read_excel("hotdog-places.xlsm")
df_hotdog_places.head()
```

```python
# importing third datasets
df_obama=pd.read_excel("obama-approval-ratings.xls")
df_obama.head()
```

```python
# Now lets create four different charts using those given datasets.
```

```python
# Bar chart.
plt.figure(figsize=(20,7))
plt.bar(df_hotdog_winner['Year'],df_hotdog_winner['Dogs eaten'],color='green')
plt.ylabel("Hotdogs Consumed")
plt.xlabel("Year")
plt.title("Hotdogs consumed Vs Year")
plt.show()
```

```python
# stacked bar chart
# Method1
df_obama.plot.bar(stacked=True)
plt.xlabel("Issues")
plt.ylabel("Rating")
plt.title("Stacked bar chart on Obama approval rating")
plt.show()


# Method2
# creating stacked bar chart using other method.
X = df_obama['Issue']
Y = df_obama['Approve']
Z= df_obama['Disapprove']
A=df_obama['None']

width = 0.30      # the width of the bars: can also be len(x) sequence
plt.figure(figsize=(25,7))

plt.bar(X, Y, width, label='Approve')
plt.bar(X, Z, width,label='disapprove',bottom=Y)
plt.bar(X, A, width,label='None')

plt.xlabel('Issues')
plt.ylabel('Obama approval rating')
plt.title("Stacked bar chart on Obama approval rating")
plt.legend()

plt.show()

```

```python
# Creating pie chart

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
X=df_obama['Issue']
Y=df_obama['Approve']
plt.figure(figsize=(25,7))

plt.pie(Y, labels=X, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.title("Pie Chart of Obama Approval rating based on Issues")

plt.show()
```

```python
# creating dunot chart
X=df_obama['Issue']
Y=df_obama['Disapprove']
plt.figure(figsize=(25,7))

plt.pie(Y, labels=X, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.title("Pie Chart of Obama Dispproval rating based on Issues")
centre_circle = plt.Circle((0,0),0.75,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```
