# Creating line and step charts using python

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing datasets
df=pd.read_excel('world-population.xlsm')
df.head()
```

```python
# lets create the line chart using matplotlib
x=df['Year']
y=df['Population']
plt.plot(x,y,color="green")
plt.title("Line Chart using Python Matplotlib")
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

```python
# creating the step charts using matplotlib
x=df['Year']
y=df['Population']
plt.plot(x, y, drawstyle='steps', linestyle='-', label='steps-pre aka steps',color="blue")
plt.plot(x, y, drawstyle='steps-mid', linestyle='--', label='steps-mid')
plt.plot(x, y, drawstyle='steps-post', linestyle=':', label='steps-post',color="red")
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Different Steps charts using Python Matplotlib")
plt.legend(loc='upper left')
plt.show()
```
