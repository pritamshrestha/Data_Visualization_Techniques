```python
# creating stacked area chart
plt.figure(figsize=(17,7))
x=Expenditure_By_Category['category'].head(10)
y=Expenditure_By_Category['expenditure'].head(10)
plt.plot(x,y,color="red")
plt.fill_between(x,y,color="blue",alpha=0.8)
plt.title("Area Chart using python",loc="center")
plt.xlabel("Category")
plt.ylabel("Expenditure")
```

```python
# creating pivot table for stacked area chart
plt_expenditures = df3.loc[:, df3.columns != 'sex'].pivot(index='year', columns='category', values='expenditure')
#print(plt_expenditures.head(1))
x=df3['year'].unique()
print(x)
a=plt_expenditures['Alcoholic Beverages']
b=plt_expenditures['Food']
c=plt_expenditures['Housing']
d=plt_expenditures['Transportation']
e=plt_expenditures['Healthcare']
f=plt_expenditures['Education']
g=plt_expenditures['Entertainment']
h=plt_expenditures['Apparel']
# plotting
plt.figure(figsize=(17,7))
plt.stackplot(x,a,b,c,d,e,f,g,h, labels=['Alcoho','Food','Housing','Trans','Health','Educat','Enter','App'])
plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Values of each category based on year")
plt.legend()
```
