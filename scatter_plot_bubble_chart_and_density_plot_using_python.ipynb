```python
# importing datasets
df2=pd.read_csv("crimerates-by-state-2005.csv")
df2.head()
```

```python
# Creating scatter plot 
plt.figure(figsize=(15,5))
#plt.scatter(df2['state'].head(13),df2['burglary'].head(13),color="green",marker='*',label="Burglary")
#plt.scatter(df2['state'].head(13),df2['motor_vehicle_theft'].head(13),color="red", marker="+",label="motor vehicle theft")
plt.scatter(df2['burglary'],df2['motor_vehicle_theft'],color='red',label="burglary vs motor theft")
plt.title("Burglary and motor vehicle theft count based on the states")
plt.xlabel("Burglary")
plt.ylabel("Motor_vehicle_theft")
plt.legend()
plt.show()
```

```python
# creating bubble plot
plt.figure(figsize=(15,5))
color=['blue','green','yellow','black','purple','pink','lightblue','orange','red','brown']
plt.scatter(df2['burglary'].head(10),df2['motor_vehicle_theft'].head(10),df2['larceny_theft']/2,c=color,label="burglary vs motor theft")
plt.title("Burglary vs motor vehicle theft count based on larceny theft")
plt.xlabel("Burglary")
plt.ylabel("Motor_vehicle_theft")
plt.legend()
plt.show()
```
